```python
for i, row in res.iterrows():
    action=''
    response=''
    if "P" in str(row['OA Vendor'])  or "I" in str(row['OA Vendor']):
        res.at[i,'Excluded reason']="Internal Vendor"
        
    else:
        
        #Lines without a response
        if pd.isna(row['Action Message Response (Select response)']) and pd.isna(row['Confirmed Delivery Date (MM/DD/YYYY) Format']):
            if (pd.isna(row['Comments (Not Mandatory)'])) or len(str(row['Comments (Not Mandatory)']))<1:
                res.at[i,'Excluded reason']="No Response"
            else:
                if str(row['AMS']).count('|')>3:
                    
                    prevAMS=str(row['AMS']).split("|")
                    res.at[i,'Action Message Status']=row['Formatted Today\'s Date'] +"|"+ "|".join(prevAMS[1:4])+'| Supplier Comment: ' + row['Comments (Not Mandatory)'] +"|"+ "|".join(prevAMS[4:])
                else:
                    res.at[i,'Action Message Status']=row['Formatted Today\'s Date'] + '| Supplier Comment: '+row['Comments (Not Mandatory)'] + '|BOT*' +"\n"+ str(row['AMS'])
                    
                
        else:
            if row['Acknowledged (Yes/No)']=='No' and ~pd.isna(row['Confirmed Delivery Date (MM/DD/YYYY) Format']):
                if row['Confirmed Delivery Date (MM/DD/YYYY) Format']== row['Delivery/Dock Date']:
                    note='Conf|On Time, Supplier Will Ship on ' + row['Confirmed Delivery Date (MM/DD/YYYY) String']
                else:
                    note='Conf|Supplier modified delivery date from {} to {}'.format(row['Delivery/Dock Date String'],row['Confirmed Delivery Date (MM/DD/YYYY) String'])
            else:
                note='DlvDateUpdated'
            if pd.isna(row['Exception Message Flag']):
                if ~pd.isna(row['Confirmed Delivery Date (MM/DD/YYYY) Format']):
                    if len(str(row['Action Message Priority R']))>15:
                        res.at[i,'Action Message Priority Response']=  '{}|{}:{}-{}'.format(row['Action Message Priority'],row['Formatted Today\'s Date'],str(row['Action Message Priority R'])[str(row['Action Message Priority R']).find(":")+1:str(row['Action Message Priority R']).find("-")],row['Confirmed Delivery Date (MM/DD/YYYY) String'])
                if row['Acknowledged (Yes/No)']=='No':
                    res.at[i,'Delivery Date']=row['Confirmed Delivery Date (MM/DD/YYYY) Format']
                    res.at[i,'Vendor Performance indicator']='X'
                    res.at[i,'Vendor Performance Date']=row['Confirmed Delivery Date (MM/DD/YYYY) Format']
                    if str(row['AMS']).count('|')>3:
                        prevAMS=str(row['AMS']).split("|")
                        res.at[i,'Action Message Status']=row['Formatted Today\'s Date'] +"|"+ "|".join(prevAMS[1:4])+'| Supplier Comment: ' + row['Comments (Not Mandatory)'] +"|"+ "|".join(prevAMS[4:])
                    else:
                        res.at[i,'Action Message Status']=row['Formatted Today\'s Date'] + '| Supplier Comment: '+row['Comments (Not Mandatory)'] + '|BOT*' +"\n"+ str(row['AMS'])
                elif row['Acknowledged (Yes/No)']=='Yes':
                    res.at[i,'Delivery Date']=row['Confirmed Delivery Date (MM/DD/YYYY) Format']
                    res.at[i,'Action Message Status']=row['Formatted Today\'s Date'] + "|ACTION|{}|BOT ACTION {}|BOT*  ".format(note,row['Comments (Not Mandatory)']) +"\n"+ str(row['AMS'])

            #Lines with AMP
            else:
                #Adding the parts of the AMP:
                if row['Exception Message Flag'] =='Expedite':
                    action='EXP'
                elif row['Exception Message Flag'] =='De-expedite':
                    action='DEX'    

                if pd.isna(row['Confirmed Delivery Date (MM/DD/YYYY) Format']) and row['Action Message Response (Select response)']!='Accept' :
                    res.at[i,'Action Message Priority Response']=  row['Action Message Priority'] + "|"+ row['Formatted Today\'s Date']+':AWRVEN'
                    res.at[i,'Delivery Date']=row['New Requested Delivery Date']
                

                if row['Action Message Response (Select response)']=='Accept':
                    response='ACC'
                    if pd.isna(row['Confirmed Delivery Date (MM/DD/YYYY)']) or row['Acknowledged (Yes/No)']=='No' :
                        res.at[i,'Action Message Priority Response']=  '{}|{}:{}{}-{}'.format(row['Action Message Priority'],row['Formatted Today\'s Date'],response,action,row['New Requested Delivery Date String'])
                        res.at[i,'Delivery Date']=row['Delivery/Dock Date']
                        res.at[i,'Vendor Performance indicator']='X'
                        res.at[i,'Vendor Performance Date']=row['New Requested Delivery Date']
                        res.at[i,'Delivery Date']=row['New Requested Delivery Date']
                    elif row['Acknowledged (Yes/No)']=='Yes':
                        res.at[i,'Action Message Priority Response']=  '{}|{}:{}{}-{}'.format(row['Action Message Priority'],row['Formatted Today\'s Date'],response,action,row['New Requested Delivery Date String'])
                        res.at[i,'Vendor Performance indicator']='X'
                        res.at[i,'Action Message Status']=row['Formatted Today\'s Date'] + "|ACTION|DlvDateUpdated|AMP BOT ACTION {}|BOT*  ".format(row['Comments (Not Mandatory)']) +"\n"+ str(row['AMS'])
                        res.at[i,'Vendor Performance Date']=row['New Requested Delivery Date']
                        res.at[i,'Delivery Date']=row['New Requested Delivery Date']
                    else:
                        res.at[i,'Excluded reason']="Not rule for that Accept"

                elif row['Action Message Response (Select response)']=='Decline' or row['Action Message Response (Select response)']=='Improved':
                    response='AT1'
                    if row['Action Message Response (Select response)']=='Improved' or row['Acknowledged (Yes/No)']=='No' :
                        res.at[i,'Delivery Date']=row['Confirmed Delivery Date (MM/DD/YYYY) Format']
                        res.at[i,'Action Message Priority Response']=  '{}|{}:{}-{}'.format(row['Action Message Priority'],row['Formatted Today\'s Date'],action+response,row['Confirmed Delivery Date (MM/DD/YYYY) String'])
                        res.at[i,'Vendor Performance indicator']='X'
                        res.at[i,'Vendor Performance Date']=row['Confirmed Delivery Date (MM/DD/YYYY) Format']
                        res.at[i,'Action Message Status']=row['Formatted Today\'s Date'] + "|ACTION|{}|BOT ACTION {}|BOT*  ".format(note,row['Comments (Not Mandatory)']) +"\n"+ str(row['AMS'])

                    elif row['Acknowledged (Yes/No)']=='Yes':
                        res.at[i,'Action Message Priority Response']=  '{}|{}:{}-{}'.format(row['Action Message Priority'],row['Formatted Today\'s Date'],action+response,row['Delivery/Dock Date String'])
                        res.at[i,'Delivery Date']=row['Delivery/Dock Date']
                        res.at[i,'Action Message Status']=row['Formatted Today\'s Date'] + "|ACTION|{}|BOT ACTION {}|BOT* ".format(note,row['Comments (Not Mandatory)']) +"\n"+ str(row['AMS'])

                elif row['Action Message Response (Select response)']=='Push Out':
                    response='DELAYD'
                    res.at[i,'Delivery Date']=row['Confirmed Delivery Date (MM/DD/YYYY) Format']
                    res.at[i,'Action Message Priority Response']=  '{}|{}:{}-{}'.format(row['Action Message Priority'],row['Formatted Today\'s Date'],response,row['Confirmed Delivery Date (MM/DD/YYYY) String'])
                    res.at[i,'Action Message Status']=row['Formatted Today\'s Date'] + "|ACTION|{}|BOT ACTION {}|BOT*  ".format(note, row['Comments (Not Mandatory)']) +"\n"+ str(row['AMS'])
                    if row['Acknowledged (Yes/No)']=='No':
                        res.at[i,'Vendor Performance indicator']='X'
                        res.at[i,'Vendor Performance Date']=row['Confirmed Delivery Date (MM/DD/YYYY) Format']

                elif row['Action Message Response (Select response)']=='Shipped':
                    response='SHIPPD'
                    res.at[i,'Delivery Date']=row['Confirmed Delivery Date (MM/DD/YYYY) Format']
                    res.at[i,'Action Message Priority Response']=  '{}|{}:{}-{}'.format(row['Action Message Priority'],row['Formatted Today\'s Date'],response,row['Confirmed Delivery Date (MM/DD/YYYY) String'])
                    if row['Acknowledged (Yes/No)']=='No':
                        res.at[i,'Vendor Performance indicator']='X'
                        res.at[i,'Vendor Performance Date']=row['Confirmed Delivery Date (MM/DD/YYYY) Format']
                        res.at[i,'Action Message Status']=row['Formatted Today\'s Date'] + "|ACTION|{}|BOT ACTION{}|BOT*  ".format(note, row['Comments (Not Mandatory)'])+"\n"+ str(row['AMS'])

                elif row['Action Message Response (Select response)']=='Delivered':
                    response='DELIVD'
                    res.at[i,'Delivery Date']=row['Confirmed Delivery Date (MM/DD/YYYY) Format']
                    res.at[i,'Action Message Priority Response']=  '{}|{}:{}-{}'.format(row['Action Message Priority'],row['Formatted Today\'s Date'],response,row['Confirmed Delivery Date (MM/DD/YYYY) String'])
                    res.at[i,'Action Message Status']=row['Formatted Today\'s Date'] + "|ACTION|{}|BOT ACTION {}|BOT*  ".format(note,row['Comments (Not Mandatory)']) +"\n"+ str(row['AMS'])
                    if row['Acknowledged (Yes/No)']=='No':
                        res.at[i,'Vendor Performance indicator']='X'
                        res.at[i,'Vendor Performance Date']=row['Confirmed Delivery Date (MM/DD/YYYY) Format']

                elif row['Action Message Response (Select response)']=='Pending Review':
                    res.at[i,'Action Message Priority Response']=row['Action Message Priority'] + "|"+ row['Formatted Today\'s Date']+':AWRVEN'
                    res.at[i,'Delivery Date']=row['Delivery/Dock Date']
                    res.at[i,'Action Message Status']=row['Formatted Today\'s Date'] + "|ACTION|{}|BOT ACTION {}|BOT*  ".format(note,row['Comments (Not Mandatory)']) +"\n"+ str(row['AMS']) 

                elif row['Action Message Response (Select response)']=='Pending SLB':
                    res.at[i,'Action Message Priority Response']=row['Action Message Priority'] + "|"+ row['Formatted Today\'s Date']+':AWRSLB'
                    res.at[i,'Delivery Date']=row['Delivery/Dock Date']
                    res.at[i,'Action Message Status']=row['Formatted Today\'s Date'] + "|ACTION|{}|BOT ACTION {}|BOT*  ".format(note,row['Comments (Not Mandatory)']) +"\n"+ str(row['AMS'])
                else:
                    res.at[i,'Excluded reason']="AMP with No Response"
        
```
